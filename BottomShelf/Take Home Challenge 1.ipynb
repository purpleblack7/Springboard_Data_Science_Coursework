```python

login_time_df = pd.DataFrame()
login_time_df = df_logins['login_time'].copy()


login_time_df = login_time_df.to_frame()
```

```python
#Extracting the date and time from the Timestamp
login_time_df['Date'] = login_time_df['login_time'].dt.date
login_time_df['Time'] = login_time_df['login_time'].dt.time
```

```python
#Setting the start and end time with a dummy date of 1-Jan-100

start_time = dt.datetime(100,1,1,0,0)
end_time = dt.datetime(100,1,1,23,59)

print(start_time)
```

```python
#Initializing a dictionary to store the bins
time_bins = {1:start_time}

time_interval = start_time
counter = 1

#Setting the delta for 15 minutes
delta = timedelta(minutes = 15)

#Creating bins for each interval of date
while time_interval< end_time:
    time_bins[counter] = time_interval
    time_interval = time_interval+ dt.timedelta(minutes = 15)
    counter+=1
    
```

```python
#Extracting only the time, thereby removing the dummy date
for k in time_bins.keys():
    time_bins[k] = time_bins[k].time()
    
```
