## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
import numpy as np
def name_finder(code):
    '''A function that finds the name for the corresponding code'''
    names = {'1':'Economic management',
             '2':'Public sector governance',
             '3':'Rule of law',
             '4':'Financial and private sector development',
             '5':'Trade and integration',
             '6':'Public sector governance',
             '7':'Social dev/gender/inclusion',
             '8':'Human development',
             '9':'Urban development',
             '10':'Rural development',
             '11':'Environment and natural resources management'
            }
    return names[code]

top_ten={}

json_df['mjtheme_namecode'].replace(' ',np.nan)
projtheme_top = json_df['mjtheme_namecode'].value_counts(['name']).index.tolist()

for each_element in projtheme_top:#selected the value in the first list, which happens to be another list
    for each_item in each_element:#selected the dictionary which lies in this nested list.
        name= each_item['name']
        if name == '' or 'nan':#check to see if the 'name' field is empty. If it is empty, the function name_finder() is called to give the suitable name for the code.
            each_item['name'] = name_finder(each_item['code'])
       
    
        if name in top_ten: #Checks if the entry is already present in a dictionary that stores the count values. If yes, the count is increased by one, otherwise a new entry is created.
            top_ten[name]+= 1
        else:
            top_ten[name] = 1
            
            
            
top_ten = pd.Series(top_ten) #converts the dictionary to Pandas Series and sorts it.
top_ten_sorted = top_ten.sort_values(ascending=False)
top_ten = top_ten_sorted.iloc[:10]
print((top_ten))
```
